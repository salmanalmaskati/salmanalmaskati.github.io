```python
import tabulate

table = [['Position', 'Team', 'Points' ,'Top Scorer']]

for team in info:
    teamm=team['team']
    pos=team['pos']
    top=team['top_scorer']
    points=team['points']
    table.append([pos, teamm, points , top])

    formatted_table = tabulate.tabulate(table, headers='firstrow', tablefmt='fancy_grid')

print(formatted_table)
```
