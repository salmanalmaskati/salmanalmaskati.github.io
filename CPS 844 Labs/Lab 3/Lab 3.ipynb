```python
page = requests.get('https://www.w3schools.com/html/html_tables.asp')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
tree = html.fromstring(page.content)
```

```python
tree.xpath('//div[@class="w3-panel w3-note"]/p/text()')
```

**Reading Header of the table**

```python
tableHeader = tree.xpath('//div[@class="w3-white w3-padding notranslate w3-padding-16"]/table[@class="ws-table-all"]/tr/th/text()')
tableHeader
```

**Reading table data**

```python
tableData = tree.xpath('//div[@class="w3-white w3-padding notranslate w3-padding-16"]/table[@class="ws-table-all"]/tr/td/text()')
tableData
```

**Preparing a dataframe**

```python
df = pd.DataFrame({tableHeader[0]: tableData[::3],
     tableHeader[1]: tableData[1::3],
     tableHeader[2]:tableData[2::3]
    })
df
```

**Write Python code for the following bullets**
* **Pick/Prepare 5 documents of your choice** (web pages/text written by you/text from anywhere)
* **Prepare vectors for these 5 documents** (Write a python code to prepare vectors)
* **Calculate cosine similarity of doc 1 with other 4 docs** (You can use any implementation of cosine similarity)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from scipy import spatial
from sklearn.feature_extraction.text import TfidfVectorizer



# Specify the URL
url1 = 'https://www.adecesg.com/resources/blog/environmental-impacts-of-renewable-energy-sources/#:~:text=Electricity%20from%20renewable%20energy%20sources,70%2D90%25%20less%20pollution.'
url2='https://www.ucsusa.org/resources/environmental-impacts-renewable-energy-technologies'
url3='https://www.energy.gov/eere/environmental-impacts-clean-energy'
url4='https://www.cbc.ca/news/canada/ottawa/bee-swarm-sparks-street-1.6896811'
url5='https://bc.ctvnews.ca/6-cars-impounded-21-tickets-issued-in-surrey-street-racing-crackdown-1.6467218'

# Fetch the HTML content 1 using requests 
response1 = requests.get(url1)
html_content1 = response1.content

# Create a Beautiful Soup 1 object
soup1 = BeautifulSoup(html_content1, 'html.parser')

# Fetch the HTML content 2 using requests
response2 = requests.get(url2)
html_content2 = response2.content

# Create a Beautiful Soup 2 object
soup2 = BeautifulSoup(html_content2, 'html.parser')

# Fetch the HTML content 3 using requests
response3 = requests.get(url3)
html_content3 = response3.content

# Create a Beautiful Soup 3 object
soup3 = BeautifulSoup(html_content3, 'html.parser')

# Fetch the HTML content 4 using requests
response4 = requests.get(url4)
html_content4 = response4.content

# Create a Beautiful Soup 4 object
soup4 = BeautifulSoup(html_content4, 'html.parser')

# Fetch the HTML content 5 using requests
response5 = requests.get(url5)
html_content5 = response5.content

# Create a Beautiful Soup 5 object
soup5 = BeautifulSoup(html_content5, 'html.parser')

#Parse url1
title1=soup1.find('title')
p_content1=''
para1=soup1.find_all('p')
for p in para1:
    para_content=p.get_text()
    p_content1+= para_content
print(f"Title 1:{title1.get_text()}")


#Parse url2
title2=soup2.find('title')
p_content2=''
para2=soup2.find_all('p')
for p in para2:
    para_content2=p.get_text()
    p_content2+= para_content2
print(f"Title 2:{title2.get_text()}")

#Parse url3
title3=soup3.find('title')
p_content3=''
para3=soup3.find_all('p')
for p in para3:
    para_content3=p.get_text()
    p_content3+= para_content3
print(f"Title 3:{title3.get_text()}")



#Parse url4
title4=soup4.find('title')
p_content4=''
para4=soup4.find_all('p')
for p in para4:
    para_content4=p.get_text()
    p_content4+= para_content4
print(f"Title 4:{title4.get_text()}")

#Parse url5
title5=soup5.find('title')
p_content5=''
para5=soup5.find_all('p')
for p in para5:
    para_content5=p.get_text()
    p_content5+= para_content5
print(f"Title 5:{title5.get_text()}")


#Term frequency
vectorizer = TfidfVectorizer()
vectorizer.fit([p_content1,p_content2,p_content3,p_content4,p_content5])

vector1 = vectorizer.transform([p_content1])
vector2 = vectorizer.transform([p_content2])
vector3 = vectorizer.transform([p_content3])
vector4 = vectorizer.transform([p_content4])
vector5 = vectorizer.transform([p_content5])

#1D Array
vector1 = vector1.toarray().flatten()
vector2 = vector2.toarray().flatten()
vector3 = vector3.toarray().flatten()
vector4 = vector4.toarray().flatten()
vector5 = vector5.toarray().flatten()

#Cosine Similarity

similarity_1_2 = 1 - spatial.distance.cosine(vector1, vector2)
print(f"\nThe cosine similairty between Title 1:Environmental Impacts of Renewable Energy and \n Title 2: Environmental Impacts of Renewable Energy Technologies is {similarity_1_2}")

similarity_1_3 = 1 - spatial.distance.cosine(vector1, vector3)
print(f"\nThe cosine similairty between Title 1:Environmental Impacts of Renewable Energy and \n Title 3: Environmental Impacts of Clean Energy | Department of Energy is {similarity_1_3}")

similarity_1_4 = 1 - spatial.distance.cosine(vector1, vector4)
print(f"\nThe cosine similairty between Title 1:Environmental Impacts of Renewable Energy and \n Title 4 30,000 bees swarm Sparks Street, break tree branch | CBC News is {similarity_1_4}")

similarity_1_5 = 1 - spatial.distance.cosine(vector1, vector5)
print(f"\nThe cosine similairty between Title 1:Environmental Impacts of Renewable Energy and \n Title 5:Surrey RCMP crackdown on street racing, stunting | CTV News is {similarity_1_5}")

#create a dictionary with key value pairs to store similarties 
my_dict = {"Document 1 and Document 2": similarity_1_2, "Document 1 and Document 3": similarity_1_3, "Document 1 and Document 4": similarity_1_4,"Document 1 and Document 5": similarity_1_5}

#print highest similarity 
print(f"\nThe document with the highest cosine similarity is {max(my_dict)} with a value of {my_dict[max(my_dict, key=my_dict.get)]} ")
```
